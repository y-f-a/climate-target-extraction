# Inspect SBTi data

Data obtained from:
- https://sciencebasedtargets.org/target-dashboard

```python
companies_file_path = DATA_DIR / "SBTi" / "companies-excel.xlsx"
targets_file_path = DATA_DIR / "SBTi" / "targets-excel.xlsx"

df_companies = pd.read_excel(companies_file_path)
df_targets = pd.read_excel(targets_file_path)

display(df_companies.head())
display(df_targets.head())
```

```python
# We only need to use the targets dataframe

sbti_ids = {
    "NVDA": 40018902,
    "MSFT": 40005011,
    "AAPL": 40003329,
    "GOOGL": 40009770,
    "AMZN": 40010768,
    "META": 40011493,
    "TSLA": 40009923,
}

# keep only mag7 companies
df_targets_filtered = df_targets[df_targets["sbti_id"].isin(sbti_ids.values())]
# keep only active targets (absolute or intensity, not renewable energy or supplier engagement targets, nor target commitments)
df_targets_filtered = df_targets_filtered[df_targets_filtered["type"].isin(["Absolute","Intensity"])]
df_targets_filtered.head()
```

```python
# get a summary
target_summary = (
    df_targets_filtered
      .groupby(["company_name","action","status"], dropna=False)
      .size()
      .reset_index(name="count")
      .sort_values("company_name", ascending=True)
)
target_summary
```
